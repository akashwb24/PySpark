```python
df = spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Products.csv").option("header",True).option("inferSchema",True).load()
display(df)
```

```python
from pyspark.sql.functions import col, when
```

```python
df2 = df.withColumn("Tier",when(col("Country")=="Australia",1))
display(df2)
```

```python
df3=df.withColumn("Tier",when(col("Country")=="Australia",1).otherwise(0))
display(df3)
```

```python
df4=df.withColumn("Tier",when(col("Country")=="Australia",1).when(col("Country")=="Canada",2).otherwise(3))
df4.display()
```

```python
df5=df.withColumn("Tier",when(col("Country")=="Australia",1).when(col("Country")=="Canada",2).when(col("Country")=="France",3).when(col("Country")=="Germany",4).otherwise(5))
display(df5)
```
