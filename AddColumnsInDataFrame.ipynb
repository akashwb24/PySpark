```python
df = spark.read.format("csv").option("path", "/Volumes/workspace/default/practice/Products.csv").option("header",True).option("inferSchema",True).load()
```

```python
df.display()
```

```python
from pyspark.sql.functions import col, round, lit
```

```python
df2 = df.withColumn("Vat",round(col("SalesAmount") * 0.05,2))
df2.display()
```

```python
len(df2.columns)
```

```python
df3 = df2.withColumn("CountryOfOrigin",lit("India")) #string cannot be given as argumnet, needs to cast into int
df3.display()
```

```python
df4  = df3.withColumn("TotalProductCost", round(col("TotalProductCost"),2))
df4.display()
```

```python
dff = spark.read.format("csv").option("path", "/Volumes/workspace/default/practice/Products.csv").option("header",True).load()
dff.display()
```

```python
df5 = dff.withColumn("ProductKey",col("ProductKey").cast("int"))
display(df5)
```

```python
df6 = dff.withColumn("ProductKey",col("ProductKey").cast("int")).withColumn("SalesAmount",col("SalesAmount").cast("double"))
display(df6)
```

```python
df7 = df.withColumns({'Vat':round(col("SalesAmount")*0.05,2)})
df7.display()
```

```python
df8 = df.withColumns({'Vat':round(col("SalesAmount")*0.05,2), 'TotalProductCost':col("TotalProductCost").cast("double")})
df8.display()
#withColumns take dictionary, we have to give key value pair
```
