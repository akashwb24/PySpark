```python
df = spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Products.csv").option("header",True).option("inferSchema",True).load()
display(df)
```

```python
from pyspark.sql.functions import *
```

```python
df2= df.select(sum("SalesAmount"))
display(df2)
```

```python
df3 = df.select(round(sum("SalesAmount"),2).alias("TotalSales"))
display(df3)
```

```python
df3 = df.select(round(avg("SalesAmount"),2).alias("AvgSales"))
display(df3)
```

```python
df3 = df.select(min("SalesAmount")).alias("MinSales")
display(df3)
```

```python
df3 = df.select(max("SalesAmount")).alias("MaxSales")
display(df3)
```

```python
df3 = df.select(count("Country")).alias("CountryCount")
display(df3)
```

```python
df3 = df.select(countDistinct("Country")).alias("CountryCount")
display(df3)
```

```python
df4= df.select(round(sum("SalesAmount"),2).alias("TOTALSALES"),round(avg("SalesAmount"),2).alias("AVGSALES"))
display(df4)
```

```python
df6 = df.selectExpr("round(sum(SalesAmount),2) as TotalSales")
display(df6)
```

```python
df6 = df.selectExpr("round(sum(SalesAmount),2) as TotalSales","round(avg(SalesAmount),2) as AvgSales")
display(df6)
```
