```python
df= spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Products.csv").option("header",True).option("inferSchema",True).load()
display(df)
```

```python
df2 = df.groupBy("Country").sum("SalesAmount")
display(df2)
```

```python
df3 = df.groupBy("Country","EnglishProductName").sum("SalesAmount")
display(df3)
```

```python
from pyspark.sql.functions import *
```

```python
df3 = df.groupBy("Country","EnglishProductName").agg(round(sum("SalesAmount"),2).alias("TotalSlaes"))
display(df3)
```

```python
df4 = df.groupBy("Country","EnglishProductName").agg(round(sum("SalesAmount"),2).alias("TotalSales"),round(avg("SalesAmount"),2).alias("AvgSales"))
display(df4)
```
