```python
df = (
    spark.read.format("csv")
    .option("path","/Volumes/workspace/default/practice/Products.csv")
    .option("header",True)
    .option("inferSchema",True)
    .load()
)
df.display()
```

```python
df.write.format("csv").option("path","/Volumes/workspace/default/practice/Bucketing/").bucketBy(5,"ProductKey").saveAsTable("bucketed_Products")
```

```python
df.write.format("csv").option("path","/Volumes/workspace/default/practice/Bucketing/").partitionBy("Country").bucketBy(5,"ProductKey").saveAsTable("bucketed2_Products")
```
