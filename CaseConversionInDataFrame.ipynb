```python
df=spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Products.csv").option("header",True).option("inferSchema",True).load()
display(df)
```

```python
from pyspark.sql.functions import col, upper, lower, initcap
```

```python
df2 = df.withColumn("Country",upper(col("Country")))
display(df2)
```

```python
df3 = df.withColumn("Country",lower(col("Country")))
display(df3)
```

```python
df4 = df.withColumn("EnglishProductName",lower(col("EnglishProductName")))
display(df4)
```

```python
df5 = df4.withColumn("EnglishProductName",initcap(col("EnglishProductName")))
display(df5)
```
