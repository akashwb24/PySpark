```python
dfchangesnew = (
    spark.read.format("delta")
    .option("readChangeFeed",True)
    .option("startingVersion",2)
    .option("endingVersion",8)
    .table("tblemp")
)
display(dfchangesnew)
```

```python
dfchangesnew = (
    spark.read.format("delta")
    .option("readChangeFeed",True)
    .option("startingTimestamp",'2026-03-31T10:08:05.000+00:00')
    .option("endingTimestamp",'2026-03-31T10:22:51.000+00:00')
    .table("tblemp")
)
display(dfchangesnew)
```
