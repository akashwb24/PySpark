```python
lst = [(1,"john",10000),(2,"sam",20000),(3,"raj",30000),(4,"raj",40000)]
print(lst)
```

```python
rdd1= spark.parallelize(lst)
rdd1.collect()
```

```python
df=rdd1.toDF()
df.display()
```

```python
df1=rdd1.toDF(["EmpID","Name","Salary"])
df1.display()
```

```python
rdd2=df.rdd
rdd2.collect()
```

```python
df2 = spark.createDataFrame(rdd1,["EmpID","Name","Salary"])
df2.display()
```
