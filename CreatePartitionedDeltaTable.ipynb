```python
dfproducts = spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Products.csv").option("header",True).option("inferSchema",True).load()
display(dfproducts)
```

```python
%sql
create table mydb.products_partitioned
(
  orderdate date,
  invoicenum string,
  productkey int,
  country string,
  englishproductname string,
  salesamount double,
  unitprice double,
  orderquantity integer,
  taxamt double,
  totalproductcost double
)
using delta
partitioned by (Country)
```

```python
%sql
describe extended mydb.products_partitioned
```

```python
dfproducts.write.format("delta").mode("append").saveAsTable("mydb.products_partitioned")
```

```python
%sql
select * from mydb.products_partitioned
```

```python
%sql
create table mydb.products_partitioned2
(
  orderdate date,
  invoicenum string,
  productkey int,
  country string,
  englishproductname string,
  salesamount double,
  unitprice double,
  orderquantity integer,
  taxamt double,
  totalproductcost double,
  orderyear int generated always as (YEAR(orderdate))
)
using delta
partitioned by (Country, orderyear)
```
