```python
from pyspark.sql.functions import col, dense_rank
from pyspark.sql import Window

data = [
    ("Anmol","HR",1000),
    ("Ujjwal","Admin",2000),
    ("Akash","HR",1500)
]
df = spark.createDataFrame(data,["name","dept","salary"])

window_spec = Window.orderBy(col("salary").desc())

# Just fetch 2nd highest salary

saldf = df.withColumn("salrank",dense_rank().over(window_spec)).filter("salrank = 2").select("name","salary")

# Fetch 2nd highest salary by department

window_s = Window.partitionBy("dept").orderBy(col("salary").desc())

salesdf = df.withColumn("salerank",dense_rank().over(window_s))\
            .filter("salerank =2") \
            .filter(col("dept") == 'HR') \
            .select("name","salary","dept")

display(salesdf)
```
