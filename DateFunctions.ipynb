```python
from pyspark.sql.functions import *
```

```python
current_date()
```

```python
df = spark.range(1,10)
df.display()
```

```python
df2=df.withColumn("CurrentDate",current_date())
df2.display()
```

```python
df3=df2.withColumn("CurrentDateTime",current_timestamp())
df3.display()
```

```python
df4=df3.withColumn("Year",year("CurrentDate")).withColumn("Month",month("CurrentDate")).withColumn("Day",dayofmonth("CurrentDate"))
df4.display()
```

```python
df5=df4.withColumn("Hour",hour("CurrentDateTime")).withColumn("Minute",minute("CurrentDateTime")).withColumn("Seconds",second("CurrentDateTime"))
df5.display()
```

```python
df6=df5.withColumn("ADD10Days",date_add("CurrentDate",10))
df6.display()
```

```python
df7=df6.withColumn("SUB11Days",date_sub("CurrentDate",10))
df7.display()
```

```python
df8=df7.withColumn("ADD3Months",add_months("CurrentDate",3))
df8.display()
```

```python
df9=df8.withColumn("DateDiff",datediff("ADD3Months","CurrentDate")) #datediff(enddate, latestdate)output is in days, 
df9.display()
```

```python
df10=df9.withColumn("CurrentDate2",date_format("CurrentDate","dd-MM-yyyy"))
display(df10)
```

```python
df11=df10.withColumn("CurrentDateTime",date_format("CurrentDateTime","dd-MM-yyyy HH:mm:ss"))
df11.display()
```
