```python
%sql
create table mydb.employees
(
  empid int, empname string,
  doj timestamp,
  yearofjoining int generated always as (YEAR(doj)),
  monthofjoining int generated always as (MONTH(doj)),
  dayofjoining int generated always as (DAY(doj))
)
using DELTA
```

```python
%sql
describe formatted mydb.employees
```

```python
%sql
insert into mydb.employees (empid, empname, doj) values (1,'john','2020-01-01'), (2,'mary','2020-02-01'), (3,'peter','2020-03-01')
```
