```python
customerdf = (
    spark.read.format("csv")
    .option("path","/Volumes/workspace/default/practice/Customers.csv")
    .option("header","true")
    .option("inferSchema","true")
    .load()
)
display(customerdf)
```

```python
from pyspark.sql.functions import udf, col
from pyspark.sql.types import StringType
import random
import string

# Define UDF to generate random email
def generate_email(customer_name):
    if customer_name is None:
        return None
    
    # Clean name → lowercase, remove spaces
    name_clean = customer_name.lower().replace(" ", ".")
    
    # Random 4 digit number suffix
    random_num = random.randint(10, 99)
    
    # Random domain
    domains = ["gmail.com", "yahoo.com", "outlook.com", "hotmail.com", "email.com"]
    domain = random.choice(domains)
    
    return f"{name_clean}{random_num}@{domain}"

# Register as UDF
email_udf = udf(generate_email, StringType())

# Add email column
df2 = customerdf.withColumn("EmailAddress", email_udf(col("CustomerName")))
df2.display()
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
import random

def generate_mobile():
    # Indian mobile format → starts with 6,7,8,9 + 9 more digits
    first_digit = random.choice(["6", "7", "8", "9"])
    remaining   = ''.join([str(random.randint(0, 9)) for _ in range(9)])
    return f"+91 {first_digit}{remaining}"

mobile_udf = udf(generate_mobile, StringType())

df3 = df2.withColumn("MobileNumber", mobile_udf())
df3.display()
```
