```python
from pyspark.sql.types import *
```

```python
mySchema = StructType((
    StructField("id",IntegerType()),
    StructField("name",StringType()),
    StructField("city",StringType()),
    StructField("salary",IntegerType())
))
df2 = spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Employees.csv").option("header",True).schema(mySchema).load()
df2.display()
```
