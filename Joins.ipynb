```python
salesdf=(
    spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Sales.csv").option("header",True).option("inferSchema",True).load()
)
display(salesdf)
```

```python
productsdf=(spark.read.format("csv").option("path","/Volumes/workspace/default/practice/ProductsNew.csv").option("header",True).option("inferSchema",True).load())
display(productsdf)
```

```python
help(salesdf.join)
```

```python
joindf=salesdf.join(productsdf)
display(joindf)
```

```python
joindf2=salesdf.join(productsdf,salesdf.ProductKey == productsdf.ProductKey)
display(joindf2)
```

```python
joindf3=salesdf.join(productsdf, salesdf.ProductKey==productsdf.ProductKey, "inner")
display(joindf3)
```

```python
joindf4=salesdf.join(productsdf, salesdf.ProductKey==productsdf.ProductKey, "left")
display(joindf4)
```

```python
joindf5=salesdf.join(productsdf, salesdf.ProductKey==productsdf.ProductKey, "right")
display(joindf5)
```

```python
joindf6=salesdf.join(productsdf, salesdf.ProductKey==productsdf.ProductKey, "leftanti")
display(joindf6)
```

```python
joindf7=salesdf.join(productsdf, salesdf.ProductKey==productsdf.ProductKey, "leftsemi")
display(joindf7)
```

```python
customerdf=(
    spark.read.format("csv")
    .option("path","/Volumes/workspace/default/practice/Customers.csv")
    .option("header",True)
    .option("inferSchema",True)
    .load()
)
display(customerdf)
```

```python
join3dfs=salesdf.join(productsdf,salesdf.ProductKey==productsdf.ProductKey,"inner").join(customerdf, salesdf.CustomerKey==customerdf.CustomerKey,"inner")
display(join3dfs)
```

```python
joindfdropdupcol=salesdf.join(productsdf,salesdf.ProductKey==productsdf.ProductKey,"inner").drop(salesdf.ProductKey)
display(joindfdropdupcol)
```
