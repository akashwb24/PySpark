```python
%sql
create table tbl_employees_stage
(
  id int,
  name string,
  city string,
  salary double
)
using delta
```

```python
%sql
copy into tbl_employees_stage
from(
  select
  id::int,
  name::string,
  city::string,
  salary::double
  from '/Volumes/workspace/default/practice/Employees.csv'
)
fileformat = csv
format_options('header' = 'true')
```

```python
%sql
merge into tbl_employee as t
using tbl_employees_stage as s
on t.id = s.id

when matched then
update set t.city = s.city,
            t.salary = s.salary
when not matched then
insert (id, name, city, salary) values (id, name, city, salary)
when not matched by source then
delete
```
