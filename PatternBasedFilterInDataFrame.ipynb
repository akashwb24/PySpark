```python
df = spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Products.csv").option("header",True).option("inferSchema",True).load()
display(df)
```

```python
from pyspark.sql.functions import col
```

```python
df2 = df.filter(col("EnglishProductName").startswith("A")) # it is case sensitive
display(df2 )
```

```python
df3 = df.filter(col("EnglishProductName").endswith("e"))
display(df3)
```

```python
df4 = df.filter(col("EnglishProductName").contains("a"))
display(df4)
```

```python
df5 = df.filter(col("EnglishProductName").like("A%")) # % means 0 to unknown of characters
display(df5)
```

```python
df6 = df.filter(col("EnglishProductName").like("%e"))
display(df6)
```

```python
df7 = df.filter(col("EnglishProductName").like("%a%"))
display(df7)
```

```python
df8 = df.filter(col("EnglishProductName").like("A_C%"))
display(df8)
```
