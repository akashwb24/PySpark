```python
df = spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Products.csv").option("header",True).option("inferSchema",True).load()
display(df)
```

```python
print(spark.version)
```

```python
df2 = df.groupBy("EnglishProductName").pivot("Country").sum("SalesAmount")
display(df2)
```

```python
df3 = df.groupBy("ProductKey","EnglishProductName").pivot("Country").sum("SalesAmount")
df3.display()
```

```python
df4 = df2.unpivot("EnglishProductName",["Australia"],"Country", "SalesAmount")
display(df4)
```

```python
df5 = df2.unpivot("EnglishProductName",["Australia","Canada"],"Country", "SalesAmount")
df5.display()
```

```python
df6 = df3.unpivot(["ProductKey","EnglishProductName"], ["Australia","Canada"],"Country","SalesAmount")
df6.display()
```

```python
df6 = df3.unpivot(["ProductKey","EnglishProductName"], ["Australia","Canada"],"Country","SalesAmount").dropna()
df6.display()
```
