```python
dfproducts = spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Products.csv").option("header",True).option("inferSchema",True).load()
display(dfproducts)
```

```python
dfproducts.write.format("parquet").option("path","/Volumes/workspace/default/practice/ProductsSE").save()
```

```python
dfnew = spark.createDataFrame([(0.05,0.01),(0.04,0.02),(0.03,0.03),(0.02,0.04)],["TaxRate","Discount"])
display(dfnew)
```

```python
dfnew.write.format("parquet").mode("append").option("path","/Volumes/workspace/default/practice/ProductsSE").save()
```

```python
df_products_se = spark.read.format("parquet").option("path","/Volumes/workspace/default/practice/ProductsSE").load()
display(df_products_se)
```

```python
dfproducts.write.format("delta").option("path","/Volumes/workspace/default/practice/ProductsDelta").save()
```

```python
dfnew.write.format("delta").mode("append").option("mergeSchema",True).option("path","/Volumes/workspace/default/practice/ProductsDelta").save()
```

```python
df_delta_se = spark.read.format("delta").option("path","/Volumes/workspace/default/practice/ProductsDelta").load()
display(df_delta_se)
```
