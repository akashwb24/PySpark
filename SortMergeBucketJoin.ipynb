```python
from pyspark.sql.functions import *
```

```python
df0=spark.range(10000000)
display(df0)
```

```python
df = spark.range(10000000).withColumn("id2", expr("id * 2"))
display(df)
```

```python
df2=df
display(df2)
```

```python
joindf = df.alias("a").join(df2.alias("b"), col("a.id") == col("b.id"))
display(joindf)
```

```python
joindf.explain(True)
```

```python
df.write.format("parquet").mode("overwrite").bucketBy(10,"id").saveAsTable("BucketIds")
```

```python
bucketdf = spark.table("BucketIds")
```

```python
joindf2 = df.join(bucketdf, df.id == bucketdf.id)
display(joindf2)
```

```python
bucketdf2 = spark.table("BucketId")
```

```python
joindf3 = bucketdf.join(bucketdf2, bucketdf.id == bucketdf2.id)
display(joindf3)
```

```python
joindf3.explain(True)
```

```python
joindf4 = bucketdf.join(bucketdf2, bucketdf.id == bucketdf2.id2)
display(joindf4)
```
