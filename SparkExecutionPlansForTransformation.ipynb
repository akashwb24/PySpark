```python
df = (spark.read.format("csv")
.option("path", "/Volumes/workspace/default/practice/Products.csv")
.option("header", True)
.option("inferschema", True)
.load())
display(df)
```

```python
df2= df.filter("Country='India'")
df2.display()
```

```python
from pyspark.sql.functions import *
```

```python
df3= df.withColumn("CountryOfOrigin",lit("India"))
df3.display()
```

```python
df4=df.withColumn("CountryOfOrigin",lit("India")).filter("Country='France'")
df4.display()
```

```python
df5=df.select(sum("SalesAmount")).alias("TotalSales")
df5.display()
```

```python
df6=df.groupBy("Country").sum("SalesAmount")
df6.display()
```

```python
countryDistinctdf=df.select("Country").distinct()
display(countryDistinctdf)
```
