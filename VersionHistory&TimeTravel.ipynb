```python
df = (
    spark.read.format("csv")
    .option("path","/Volumes/workspace/default/practice/Products.csv")
    .option("header",True)
    .option("inferSchema",True)
    .load()
)
display(df)
```

```python
df.write.format("delta").saveAsTable("tblProducts")
```
