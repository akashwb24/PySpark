```python
df = spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Employees.csv").option("header",True).option("inferSchema",True).load()
df.display()
```

```python
from pyspark.sql.window import *
from pyspark.sql.functions import *
```

```python
mywin = Window.orderBy("Salary")
df2 = df.withColumn("RowNum",row_number().over(mywin)) #default is ascending
df2.display()
```

```python
mywin = Window.orderBy(col("Salary").desc())
df3 = df.withColumn("RowNum",row_number().over(mywin))
display(df3)
```

```python
mywin = Window.partitionBy("City").orderBy(col("Salary"))
df4 = df.withColumn("RowNum",row_number().over(mywin))
df4.display()
```

```python
mywin = Window.orderBy(col("Salary").desc())
df5 = df.withColumn("rowNum",rank().over(mywin))
display(df5)
```

```python
mywin = Window.orderBy(col("Salary").desc())
df5 = df.withColumn("rowNum",rank().over(mywin))
display(df5)
```

```python
mywin = Window.orderBy(col("Salary"))
df6=df.withColumn("CumuTotal",sum("Salary").over(mywin))
df6.display()
```

```python
mywin=Window.orderBy(col("Salary")).rowsBetween(Window.unboundedPreceding,Window.currentRow)
df7=df.withColumn("Cumulative Salary",sum(col("Salary")).over(mywin))
display(df7)
```

```python
mywin=Window.orderBy(col("Salary")).rowsBetween(-2,Window.currentRow)
df8=df.withColumn("Cumulative Salary",sum(col("Salary")).over(mywin))
display(df8)
```
