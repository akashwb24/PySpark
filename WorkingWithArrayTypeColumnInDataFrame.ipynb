```python
df = spark.read.format("csv").option("path","/Volumes/workspace/default/practice/Products.csv").option("header",True).option("inferSchema",True).load()
df.display()
```

```python
from pyspark.sql.functions import *
```

```python
df2 = df.select(split("EnglishProductName"," ").alias("ProductName"))#when splitting it becomes array
df2.display()
```

```python
df3=df2.selectExpr("ProductName[0]")
df3.display()
```

```python
df3=df2.selectExpr("ProductName[1]")
df3.display()
```

```python
df4=df2.withColumn("NoOfElements",size("ProductName"))
df4.display()
```

```python
df5=df2.withColumn("Condition",array_contains("ProductName","Bike"))
df5.display()
```
